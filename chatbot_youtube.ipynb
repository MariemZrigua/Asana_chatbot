## Nettoyage et prétraitement du texte

```python
pip install nltk
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
```

```python
def nettoyer_texte(texte):
    # Suppression des caractères non alphabétiques
    texte = re.sub(r'[^a-zA-Z\s]', '', texte)

    # Conversion en minuscules
    texte = texte.lower()

    # Tokenisation
    mots = word_tokenize(texte)

    # Suppression des stopwords
    mots = [mot for mot in mots if mot not in stopwords.words('french')]

    # Lemmatisation
    mots = [lemmatizer.lemmatize(token) for token in mots]

    return ' '.join(mots)
```

```python
df['texte_prétraité'] = df['Texte extrait'].apply(nettoyer_texte)
```
